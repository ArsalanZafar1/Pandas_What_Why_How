What is Pandas?

Pandas is a fast, powerful, easy-to-use open-source python library for data manupulation and analysis. It is built on top of  NumPy (another python library)which is a multidimensional array with a only homogeneous elements. While Pandas is designed to work with **heterogeneous data** with rows and columns. It works with tabular, textual, also time-series data. Its function and structure are similar to an spreadsheet or SQL tables.

Pandas has two data structures, **Series** and **Dataframe** which we will review in detail. 

Why should we learn Pandas?

Pandas is fast , flexible and effcient in analyzing data from various sources and it is open source. Pandas can be run on your desktop and in the cloud for faster computing power. It is the top choice for data scientists and there is a large Pandas community to lean on. Pandas was born from python and this aspect gives it a massive reach into the type of data sources and capabilities you can build around your data analysis.

Data Structures in Pandas
Pandas uses two data structures Series and Data frame to store and analyze data.  

**Series**

A Series is a one-dimensional array with a single column that can hold hetrogenous data (integers, strings, floating point numbers, Python objects, etc.).It has an index that can be used to call different data values. elements. In order to create a Series in pandas, first, we need to import pandas library.

Series function in its basic form only needs a list of data to be created. 

pd.Series(data)


```python
data = [2,3,4,5,6]
series_1 = pd.Series(data)
series_1
```

The output shows index on the left axis (numbers numbers from 0 to 4) and data in the column on the right.

Lets name our series

```python
series_1 = pd.Series(data = [2,3,4,5,6], name='hours' )

series_1
```

And finally lets set our our own index values.

```python
sale = pd.Series([500,85000,40000,9600,100000],index = ['Cookie', 'Choco', 'Toothpaste', 'Soap', 'Toffee'])
sale
```

**DataFrame**

Dataframe is a two dimensinal array. It can also be thought of as a two or more Series. It looks like a spreadsheet or tabular data with rows, columns and an index. It has two acis, axis 0 and axis 1. 0 always means rows, and 1 means columns.

The most straightforward way to create a dataframe is from the dictionary with keys and corespnsing lists. Index by default is numbers (0,1,2..n) unless index defines explicitly.

Lets create a dictionary first.*italicized text*

```python
data = {'country': ['Canada', 'Canada', 'US', 'US', 'UK', 'UK'],
'year': [2010, 2011, 2012, 2013, 2014, 2015],
'pop': [2.5, 1.7, 2.6, 1.4, 4.9, 3.7]}

data
```

```python
pd.DataFrame(data)
```

DataFrame Function also takes parameters in order to create a dataframe.
- **data**: In this parameter, we assign a dict.
- **index**: It is similar to Series index (0,1,2,3.....n)
- **columns**: If not given, it assigned by default (0,1,2,3.....n)


Here just variable name is written, not a **data** parameter because DataFrame considers the first value as a data parameter. However, we have to write other parameters names like columns, index.

We can also specify the order of columns.

```python
pd.DataFrame(data, columns= ['year', 'country', 'pop'])
```

If we create a column with no values pandas put NaN values in it. 

```python
data = pd.DataFrame(data, columns= ['year', 'country', 'pop', 'uni'])

data
```
